```python
import zipfile
import os

base_path = "data/text_files"

for root, dirs, files in os.walk(base_path):
    for file in files:
        if file.endswith(".zip"):
            zip_path = os.path.join(root, file)
            extract_folder = os.path.join(root, file.replace(".zip", ""))
            os.makedirs(extract_folder, exist_ok=True)
            
            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                zip_ref.extractall(extract_folder)

print("Unzipping complete.")
```

```python
all_txt_files = []

for root, dirs, files in os.walk(base_path):
    for file in files:
        if file.endswith(".txt"):
            all_txt_files.append(os.path.join(root, file))

print("Number of text files:", len(all_txt_files))
print("Example files:", all_txt_files[:5])
```

```python
print(os.path.basename(all_txt_files[0]))
```

```python
text_dict = {}

for file_path in all_txt_files:
    file_name = os.path.basename(file_path)
    file_id = file_name.replace(".txt", "")
    
    with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
        text_dict[file_id] = f.read()

print("Texts loaded:", len(text_dict))
```

```python
import pandas as pd

df = pd.read_csv("data/archelect_search.csv", low_memory=False)

print("Metadata rows:", len(df))
```

```python
df["text"] = df["id"].map(text_dict)

print("After merging:")
print("Missing texts:", df["text"].isna().sum())
```

```python
df = df[~df["text"].isna()].copy()

print("Remaining rows:", len(df))
```

```python
df["word_count"] = df["text"].str.split().str.len()

print(df["word_count"].describe())
```

```python
df = df[df["word_count"] > 100]

print("After length filtering:", len(df))
```

```python
party_counts = df_reg["titulaire-liste"].value_counts()

major_parties = party_counts[party_counts > 100].index

df_reg["party_clean"] = df_reg["titulaire-liste"].where(
    df_reg["titulaire-liste"].isin(major_parties),
    "Other"
)

print(df_reg["party_clean"].value_counts())
```

```python
import pandas as pd

def categorize_profession(prof):
    prof = prof.lower()

    if any(w in prof for w in ["professeur", "enseignant", "instituteur"]):
        return "Education"
    elif any(w in prof for w in ["avocat", "juriste"]):
        return "Law"
    elif any(w in prof for w in ["médecin", "docteur", "infirmier", "pharmacien"]):
        return "Health"
    elif any(w in prof for w in ["agric", "cultivat", "éleveur"]):
        return "Agriculture"
    elif any(w in prof for w in ["chef d", "entrepreneur", "commer", "gérant"]):
        return "Business"
    elif any(w in prof for w in ["ingénieur", "technicien"]):
        return "Technical"
    elif any(w in prof for w in ["ouvrier", "employé"]):
        return "Worker"
    elif any(w in prof for w in ["journaliste"]):
        return "Media"
    elif any(w in prof for w in ["maire", "député"]):
        return "Political"
    else:
        return "Other"

df_reg["profession_clean"] = df_reg["titulaire-profession"].apply(categorize_profession)

print(df_reg["profession_clean"].value_counts())
```
